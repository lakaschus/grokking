```python
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
import json

# Your data
with open('../train_data.json', 'r') as file:
        data = json.load(file)
        numbers = data["labels"]

# Calculate frequencies
counter = Counter(numbers)
values = list(counter.keys())
frequencies = list(counter.values())

# Create the plot
plt.figure(figsize=(12, 6))

# Create bar plot
plt.bar(values, frequencies, color='skyblue', edgecolor='black')

# Customize the plot
plt.title('Distribution of Numbers', fontsize=14, pad=20)
plt.xlabel('Number', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add some padding to x-axis
plt.margins(x=0.01)

# Rotate x-axis labels for better readability if needed
# plt.xticks(rotation=45)

# Add a tight layout to prevent label cutoff
plt.tight_layout()

# Optional: Add statistical information
mean_val = np.mean(numbers)
median_val = np.median(numbers)
std_val = np.std(numbers)

stats_text = f'Mean: {mean_val:.2f}\nMedian: {median_val:.2f}\nStd Dev: {std_val:.2f}'
plt.text(0.95, 0.95, stats_text,
         transform=plt.gca().transAxes,
         verticalalignment='top',
         horizontalalignment='right',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# Show the plot
plt.show()

# Print some basic statistics
print("\nBasic Statistics:")
print(f"Most common number(s): {counter.most_common(3)}")
print(f"Number of unique values: {len(counter)}")
print(f"Total numbers: {len(numbers)}")
```
